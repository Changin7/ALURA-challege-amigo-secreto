```python
import pandas as pd

# Asumiendo que store1, store2, store3, y store4 ya fueron creadas en el código base
tiendas = [store1, store2, store3, store4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Diccionario para almacenar los resultados y facilitar la visualización posterior
resultados = {}

for i, df in enumerate(tiendas):
    nombre = nombres_tiendas[i]
    print(f"--- Análisis para la {nombre} ---")

    # 1. Ingreso total de cada tienda
    ingreso_total = df['Precio'].sum()
    print(f"1. Ingreso Total: ${ingreso_total:,.2f}")

    # 2. Cantidad de productos vendidos por categoría
    ventas_categoria = df['Categoría del Producto'].value_counts()
    categoria_top = ventas_categoria.idxmax()
    print(f"2. Categoría más popular: {categoria_top} ({ventas_categoria.max()} ventas)")

    # 3. Calificaciones promedio
    calificacion_promedio = df['Calificación'].mean()
    print(f"3. Calificación promedio: {calificacion_promedio:.2f} estrellas")

    # 4. Productos más y menos vendidos
    ventas_producto = df['Producto'].value_counts()
    producto_mas_vendido = ventas_producto.idxmax()
    producto_menos_vendido = ventas_producto.idxmin()
    print(f"4. Producto más vendido: {producto_mas_vendido} ({ventas_producto.max()} unidades)")
    print(f"4. Producto menos vendido: {producto_menos_vendido} ({ventas_producto.min()} unidades)")

    # 5. Costo de envío promedio
    envio_promedio = df['Costo de envío'].mean()
    print(f"5. Costo de envío promedio: ${envio_promedio:,.2f}\n")

    # Guardamos los datos para los gráficos
    resultados[nombre] = {
        'ingreso_total': ingreso_total,
        'calificacion': calificacion_promedio,
        'envio_promedio': envio_promedio
    }
```

```python
import matplotlib.pyplot as plt

# Extraer datos del diccionario de resultados
nombres = list(resultados.keys())
ingresos = [resultados[t]['ingreso_total'] for t in nombres]
calificaciones = [resultados[t]['calificacion'] for t in nombres]
envios = [resultados[t]['envio_promedio'] for t in nombres]

# Configurar el tamaño de la figura principal
plt.figure(figsize=(18, 5))

# --- GRÁFICO 1: Ingresos Totales (Gráfico de Barras) ---
plt.subplot(1, 3, 1)
colores = ['#4C72B0', '#55A868', '#C44E52', '#8172B2']
plt.bar(nombres, ingresos, color=colores)
plt.title('Ingreso Total por Tienda', fontsize=14, fontweight='bold')
plt.ylabel('Ingresos ($)', fontsize=12)
# Formatear el eje Y para que se lea mejor
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# --- GRÁFICO 2: Calificación Promedio (Gráfico de Líneas con Marcadores) ---
plt.subplot(1, 3, 2)
plt.plot(nombres, calificaciones, color='orange', marker='o', linestyle='-', linewidth=2, markersize=8)
plt.title('Calificación Promedio de Clientes', fontsize=14, fontweight='bold')
plt.ylabel('Estrellas (1 - 5)', fontsize=12)
plt.ylim(3.5, 4.5) # Ajustamos la escala para ver diferencias

# --- GRÁFICO 3: Categorías de la Tienda 1 (Gráfico Circular / Pie Chart) ---
plt.subplot(1, 3, 3)
categorias_t1 = store1['Categoría del Producto'].value_counts().head(5) # Top 5 categorías
plt.pie(categorias_t1, labels=categorias_t1.index, autopct='%1.1f%%', startangle=90, cmap='Pastel1')
plt.title('Top 5 Categorías - Tienda 1', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('graficos_alura_store.png') # Se guarda la imagen
```
